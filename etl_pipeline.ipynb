```python
import pandas as pd

url_clientes = "https://raw.githubusercontent.com/KevinCallejas93/etl-seguros-pipeline/main/data/raw/clientes.csv"
url_aseguradoras = "https://raw.githubusercontent.com/KevinCallejas93/etl-seguros-pipeline/main/data/raw/aseguradoras.csv"
url_corredores = "https://raw.githubusercontent.com/KevinCallejas93/etl-seguros-pipeline/main/data/raw/corredores.csv"
url_tipos = "https://raw.githubusercontent.com/KevinCallejas93/etl-seguros-pipeline/main/data/raw/tipos_seguro.csv"
url_polizas = "https://raw.githubusercontent.com/KevinCallejas93/etl-seguros-pipeline/main/data/raw/polizas.csv"
url_siniestros = "https://raw.githubusercontent.com/KevinCallejas93/etl-seguros-pipeline/main/data/raw/siniestros.csv"

clientes = pd.read_csv(url_clientes)
aseguradoras = pd.read_csv(url_aseguradoras)
corredores = pd.read_csv(url_corredores)
tipos_seguro = pd.read_csv(url_tipos)
polizas = pd.read_csv(url_polizas)
siniestros = pd.read_csv(url_siniestros)
```

```python
# Exploración inicial de los datasets
for nombre, df in {
    "clientes": clientes,
    "aseguradoras": aseguradoras,
    "corredores": corredores,
    "tipos_seguro": tipos_seguro,
    "polizas": polizas,
    "siniestros": siniestros
}.items():
    print(f"\n--- Exploración de {nombre} ---")
    print("Dimensiones:", df.shape)          # filas y columnas
    print("\nInfo:")
    print(df.info())                         # tipos de datos y nulos
    print("\nPrimeras filas:")
    print(df.head())                         # primeras 5 filas
    print("\nValores nulos por columna:")
    print(df.isnull().sum())                 # conteo de nulos
    print("\nDuplicados:", df.duplicated().sum())  # número de filas duplicadas
```

```python
# Función de limpieza general
def limpiar_dataframe(df):
    # 1. Normalizar nombres de columnas
    df.columns = (
        df.columns
        .str.strip()                # quitar espacios al inicio/fin
        .str.lower()                # pasar a minúsculas
        .str.replace(" ", "_")      # reemplazar espacios por guiones bajos
    )

    # 2. Eliminar filas duplicadas
    df = df.drop_duplicates()

    # 3. Quitar espacios en valores de texto
    df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

    return df

# Aplicar la limpieza a todos los datasets
clientes = limpiar_dataframe(clientes)
aseguradoras = limpiar_dataframe(aseguradoras)
corredores = limpiar_dataframe(corredores)
tipos_seguro = limpiar_dataframe(tipos_seguro)
polizas = limpiar_dataframe(polizas)
siniestros = limpiar_dataframe(siniestros)

# Verificar que se aplicó la limpieza
for nombre, df in {
    "clientes": clientes,
    "aseguradoras": aseguradoras,
    "corredores": corredores,
    "tipos_seguro": tipos_seguro,
    "polizas": polizas,
    "siniestros": siniestros
}.items():
    print(f"\n--- {nombre} ---")
    print("Columnas:", df.columns.tolist())
    print("Duplicados:", df.duplicated().sum())
```

```python
def limpiar_dataframe(df):
    # Normalizar nombres de columnas
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
    )

    # Eliminar duplicados
    df = df.drop_duplicates()

    # Quitar espacios en valores de texto (columna por columna)
    for col in df.select_dtypes(include="object").columns:
        df[col] = df[col].map(lambda x: x.strip() if isinstance(x, str) else x)

    return df
```

```python
# Unir clientes con pólizas
clientes_polizas = pd.merge(
    clientes,
    polizas,
    on="id_cliente",
    how="inner"
)

# Unir con corredores
clientes_polizas_corredores = pd.merge(
    clientes_polizas,
    corredores,
    on="id_corredor",
    how="inner"
)

# Unir con aseguradoras
clientes_polizas_completo = pd.merge(
    clientes_polizas_corredores,
    aseguradoras,
    on="id_aseguradora",
    how="inner"
)

# Unir con tipos de seguro
clientes_polizas_completo = pd.merge(
    clientes_polizas_completo,
    tipos_seguro,
    on="id_tipo_seguro",
    how="inner"
)

# Finalmente unir con siniestros
clientes_polizas_siniestros = pd.merge(
    clientes_polizas_completo,
    siniestros,
    on="id_poliza",
    how="left"   # left para que aparezcan pólizas aunque no tengan siniestros
)

# Ver resultado
print(clientes_polizas_siniestros.head())
print("Dimensiones finales:", clientes_polizas_siniestros.shape)
```

```python
import os

# Crear las carpetas si no existen
os.makedirs("data/curated", exist_ok=True)
os.makedirs("data/rejects", exist_ok=True)

# Guardar dataset integrado en curated
clientes_polizas_siniestros.to_csv("data/curated/clientes_polizas_siniestros.csv", index=False)

# Crear dataset de rejects (filas con NaN en columnas clave)
rejects = clientes_polizas_siniestros[
    clientes_polizas_siniestros[["id_cliente", "id_poliza", "id_corredor", "id_aseguradora", "id_tipo_seguro"]].isnull().any(axis=1)
]
rejects.to_csv("data/rejects/rejects.csv", index=False)

print("Archivos guardados en data/curated y data/rejects")
```
